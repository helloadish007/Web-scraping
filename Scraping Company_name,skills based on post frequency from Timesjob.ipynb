```python
import time
import requests
from bs4 import BeautifulSoup
def find_jobs():
    html_text=requests.get('https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=python&txtLocation=').text
    soup=BeautifulSoup(html_text,'lxml')
    jobs=soup.find_all('li',class_='clearfix job-bx wht-shd-bx')
    print("Enter your unfamilar skill ")
    unfamiliar_skill=input(">")
    print(f"Filtering out {unfamiliar_skill}")
    for job in jobs:
        company_name=job.find('h3',class_='joblist-comp-name').text.replace('\n','').replace('\r','').replace(' ','')
        job_description=job.find('span',class_='srp-skills').text.replace('\n','').replace('\r','').replace(' ','')
        published_date=job.find('span',class_='sim-posted').span.text
        link=job.header.h2.a['href']
        if 'few' in published_date:
            if unfamiliar_skill not in job_description:
                print(f"COMPANY NAME : {company_name.strip()}")
                print(f"JOB_DESCRIPTION : {job_description.strip()}")
                print(f"link: {link}")
                print(" ")

if __name__=='__main__': # checks if we are running the current file only and not an imported one
    while True:
        find_jobs()
        time_wait=2
        print(f"waiting time is {time_wait} minutes")
        time.sleep(time_wait*60)
        
```
